Let's organize the data into two variables: advertising expenditure (independent variable) and sales (dependent variable).

Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]

Sales (in thousands): [25, 30, 28, 20, 26]

Step 2: Calculate the mean of each variable

We need to calculate the mean of the advertising expenditure and sales.

Mean of Advertising Expenditure:
    
(10 + 15 + 12 + 8 + 14) / 5 = 11.8

Mean of Sales:
(25 + 30 + 28 + 20 + 26) / 5 = 25.8

Step 3: Calculate the deviation of each variable from the mean

For each value in the variables, we calculate the deviation from the mean.

Deviation of Advertising Expenditure = Advertising Expenditure - Mean of Advertising Expenditure

Deviation of Sales = Sales - Mean of Sales

Deviation of Advertising Expenditure: [-1.8, 3.2, 0.2, -3.8, 2.2]

Deviation of Sales: [-0.8, 4.2, 2.2, -5.8, 0.2]

Step 4: Calculate the squared deviation of each variable

We calculate the squared deviation for each value in the variables.

Squared Deviation of Advertising Expenditure = (Deviation of Advertising Expenditure)²

Squared Deviation of Sales = (Deviation of Sales)²

Squared Deviation of Advertising Expenditure: [3.24, 10.24, 0.04, 14.44, 4.84]

Squared Deviation of Sales: [0.64, 17.64, 4.84, 33.64, 0.04]

Step 5: Calculate the product of the deviations for each observation

We multiply the deviation of advertising expenditure and sales for each observation.

Product of Deviations = Deviation of Advertising Expenditure * Deviation of Sales

Product of Deviations: [1.44, 13.44, 0.44, 22.04, 0.44]

Step 6: Calculate the sum of squared deviations and the sum of the product of deviations

We calculate the sum of squared deviations of advertising expenditure, the sum of squared deviations of sales, and the sum of the product of deviations.

Sum of Squared Deviations of Advertising Expenditure: 32.8

Sum of Squared Deviations of Sales: 56.8

Sum of Product of Deviations: 37.8

Step 7: Calculate the regression coefficients (b1 and b0)

To calculate the regression coefficients, we use the following formulas:

b1 = (Sum of Product of Deviations) / (Sum of Squared Deviations of Advertising Expenditure)

b0 = (Mean of Sales) - (b1 * Mean of Advertising Expenditure)

b1 = 37.8 / 32.8 ≈ 1.1524

b0 = 25.8 - (1.1524 * 11.8) ≈ 12.91

Step 8: Write the regression equation

The regression equation is given by:

Sales = b0 + b1 * Advertising Expenditure

Substituting the values:

Sales = 12.91 + 1.1524 * Advertising Expenditure

So, the multiple regression equation to predict sales based on advertising expenditure is:

Sales = 12.91 + 1.1524 * Advertising Expenditure
